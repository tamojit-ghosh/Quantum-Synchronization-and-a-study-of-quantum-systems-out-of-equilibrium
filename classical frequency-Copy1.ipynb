```python
from scipy import linspace
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#def vdp(t, z,q0,omega):
def vdp(t,z):
    x, y = z
    return [omega*y, -mu*y*(x**2-q0**2)-omega*x]

def root(t, y): return y[0]
```

```python
mu=0.001
styles = ["-", "--", ":"]
t = np.linspace(0, 100, 10000)
q0=3
omega=1
```

```python

#sol = solve_ivp(vdp, [0, 100], [1, 0], t_eval=t,args=(myu,q0,omega),events=root)
sol = solve_ivp(vdp, [0, 100], [1, 0], t_eval=t,events=root)
plt.plot(sol.y[0], sol.y[1])
```

```python
zeros = sol.t_events[0]
```

```python
spacing = zeros[1:] - zeros[:-1]
deltas = spacing[1:] - spacing[:-1]
```
